```python
from agno.agent import Agent
from agno.models.together import Together
```

```python
reasoning_agent = Agent(
    model=Together(id="meta-llama/Llama-4-Scout-17B-16E-Instruct"),
    reasoning=True, markdown=True,
)
```

```python
reasoning_agent.print_response(
    "Solve The Lifeboat Dilemma problem. "
    "Context: A ship has sunk, and survivors are on a lifeboat built to carry 10 people. There are currently 15 people aboard and the boat is sinking under the weight. "
    "Evaluate multiple ethical frameworks. "
    "Include an ASCII diagram of your solution.",
    stream=True,
    show_full_reasoning=True,
)
```
